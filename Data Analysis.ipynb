Exploratory data analysis

```python
ax=sns.countplot(x='StoreType',data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns_data=df.groupby(['StoreType'],as_index=False)['TotalSales'].sum().sort_values(by='TotalSales',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='StoreType',y='TotalSales',data=sns_data)In
```

```python
ax=sns.countplot(x='StoreName',data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns_data=df.groupby(['StoreName'],as_index=False)['TotalSales'].sum().sort_values(by='TotalSales',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='StoreName',y='TotalSales',data=sns_data)
```

```python
sns_data=df.groupby(['Location'],as_index=False)['TotalSales'].sum().sort_values(by='TotalSales',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Location',y='TotalSales',data=sns_data)
```

Insights
- In Apparel the sales is more
- In super bazar,super market the sales is less
